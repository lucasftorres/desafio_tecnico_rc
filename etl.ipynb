# Transformação

## 1    -   Tratamentos Gerais

### Conversão de datas

```python
pedidos['Data Pedido'] = pd.to_datetime(pedidos['Data Pedido'], dayfirst=True)
pedidos['Data Envio'] = pd.to_datetime(pedidos['Data Envio'], dayfirst=True)
pedidos
```

### Renomear colunas

```python
vendas = vendas.rename(columns={'Pedido': 'id_pedido',
                                 'Cliente': 'id_cliente',
                                 'Produto': 'id_produto',
                                 'Valor Venda': 'valor_venda',
                                 'Quantidade Vendida': 'qtd_vendida',
                                 'Custo Envio': 'custo_envio'})

pedidos = pedidos.rename(columns={'Local Pedido': 'local_pedido',
                                  'Ano': 'ano',
                                  'ID Pedido': 'pedido_num',
                                   'Data Pedido': 'data_pedido',
                                   'Data Envio': 'data_envio',
                                   'Modo Envio': 'modo_envio',
                                   'Prioridade Pedido': 'prioridade'})

clientes = clientes.rename(columns={'ID Cliente': 'id_cliente',
                                    'Nome Cliente': 'nome_cliente',
                                    'Segmento': 'segmento',
                                    'Cidade': 'cidade',
                                    'Estado': 'estado',
                                    'Pais': 'pais',
                                    'Mercado': 'mercado',
                                    'Regiao': 'regiao'})

produtos = produtos.rename(columns={'ID Produto': 'id_produto',
                                    'Categoria': 'categoria',
                                    'SubCategoria': 'subcategoria',
                                    'Nome Produto': 'nome_produto'})
```

### Conversão de valores

```python
vendas['valor_venda'] = vendas['valor_venda'].str.replace(',', '.').astype(float)
vendas['custo_envio'] = vendas['custo_envio'].str.replace(',', '.').astype(float)
```

### Crianção de novas colunas preco_unitario na tabela vendas e id_pedido na tabela pedidos

```python
vendas['valor_unit'] = vendas['valor_venda'] / vendas['qtd_vendida']
pedidos['id_pedido'] = pedidos['local_pedido'].astype(str) + '-' + pedidos['ano'].astype(str) + '-' + pedidos['pedido_num'].astype(str)
```

## 2    -   Tabela dim_produtos

### Contar quantidade de ocorrências por nome_produto > 1

```python
contagem = produtos['nome_produto'].value_counts().reset_index()
contagem.columns = ['nome_produto', 'Quantidade']
num_produtos_repetidos = contagem[contagem['Quantidade'] > 1]
num_produtos_repetidos
```

### Filtrando no dataframe original os produtos que estão repetidos

```python
produtos_repetidos = produtos[produtos['nome_produto'].isin(num_produtos_repetidos['nome_produto'])]
produtos_repetidos
```

```python
produtos[produtos['nome_produto'] == 'Novimex Executive Leather Armchair, Black']
```

### Identificar o id_produto mais frequente para cada nome_produto

```python
id_mais_frequente = produtos_repetidos.groupby('nome_produto')['id_produto'] \
    .agg(lambda x: x.mode().iloc[0]) \
    .reset_index()
id_mais_frequente
```

### Dicionário de mapeamento nome_produto → id_produto mais frequente

```python
mapa_id_padronizado = dict(zip(id_mais_frequente['nome_produto'], id_mais_frequente['id_produto']))
mapa_id_padronizado
```

### Criar nova coluna com id_produto padronizado

```python
produtos['id_produto_padronizado'] = produtos.apply(
    lambda row: mapa_id_padronizado.get(row['nome_produto'], row['id_produto']), axis=1
)
produtos
```

### Merge com a tabela vendas para associar os id's padronizados

```python
vendas_com_id = vendas.merge(
    produtos[['id_produto', 'id_produto_padronizado']],
    on='id_produto',
    how='left'
)
vendas_com_id.head(5)
```

### Checagem comparativa

```python
vendas_com_id[vendas_com_id['id_produto'] != vendas_com_id['id_produto_padronizado']]
```

```python
produtos.shape
```

```python
dim_produtos = produtos[produtos['id_produto'] == produtos['id_produto_padronizado']].copy()
dim_produtos.shape
```

### Criação de Surrogate Key para a tabela produtos

```python
sk_produto = {nome: idx for idx, nome in enumerate(dim_produtos['nome_produto'].unique(), 1)}
dim_produtos['sk_produto'] = dim_produtos['nome_produto'].map(sk_produto)
dim_produtos
```

### Merge tabela produtos x vendas para inserir sk_produto

```python
vendas_sk_produto = vendas_com_id.merge(
    dim_produtos[['id_produto_padronizado', 'sk_produto']],
    on='id_produto_padronizado',
    how='left'
)
vendas_sk_produto.head(10)
```

## 3    -   Tabela dim_clientes

### Verificação de clientes duplicados a partir do nome

```python
contagem = clientes['nome_cliente'].value_counts().reset_index()
contagem.columns = ['nome_cliente', 'Quantidade']
num_clientes_repetidos = contagem[contagem['Quantidade'] > 1]
num_clientes_repetidos
```

### Verificação a partir dos outros atributos

```python
cidades_por_nome = clientes.groupby('nome_cliente')['cidade'].nunique().reset_index()
cidades_por_nome = cidades_por_nome[cidades_por_nome['cidade'] > 1]

estados_por_nome = clientes.groupby('nome_cliente')['estado'].nunique().reset_index()
estados_por_nome = estados_por_nome[estados_por_nome['estado'] > 1]

regioes_por_nome = clientes.groupby('nome_cliente')['regiao'].nunique().reset_index()
regioes_por_nome = regioes_por_nome[regioes_por_nome['regiao'] > 1]

# Descobrir todos nomes com duplicidade em cidade
duplicidades_nome_cidade = clientes.duplicated(subset=['nome_cliente', 'cidade'], keep=False)
# Mas o que você quer é: mesmo nome, cidades diferentes
nomes_varias_cidades = clientes.groupby('nome_cliente')['cidade'].nunique().reset_index()
nomes_varias_cidades = nomes_varias_cidades[nomes_varias_cidades['cidade'] > 1]

# Filtrar todos os registros de nomes com cidades diferentes
clientes_nomes_varias_cidades = clientes[clientes['nome_cliente'].isin(nomes_varias_cidades['nome_cliente'])].sort_values('nome_cliente')
clientes_nomes_varias_cidades
```

### Criação da Surrogate key para tabela clientes

```python
dim_clientes = clientes.copy()
sk_cliente = {cliente: idx for idx, cliente in enumerate(clientes['id_cliente'].unique(), 1)}
dim_clientes['sk_cliente'] = clientes['id_cliente'].map(sk_cliente)
dim_clientes
```

### Merge tabela clientes x vendas_sk_produto para inserir sk_clientes

```python
vendas_sk_clientes = vendas_sk_produto.merge(
    dim_clientes[['id_cliente', 'sk_cliente']],
    on='id_cliente',
    how='left'
)
vendas_sk_clientes.head(10)
```

## 4    -   Tabela dim_pedidos

### Checagem pedidos duplicados

```python
pedido_duplicado = pedidos['pedido_num'].value_counts()
pedido_duplicado = pedido_duplicado[pedido_duplicado > 1].index
pedidos_duplicados = pedidos[pedidos['pedido_num'].isin(pedido_duplicado)]

pedidos_duplicados.sort_values(by=['pedido_num'])
```

### Criação da Surrogate key para tabela pedidos

```python
dim_pedidos = pedidos.copy()
sk_pedido = {pedido: idx for idx, pedido in enumerate(pedidos['id_pedido'].unique(), 1)}
dim_pedidos['sk_pedido'] = pedidos['id_pedido'].map(sk_pedido)
dim_pedidos
```

### Merge tabela clientes x vendas_sk_clientes para inserir sk_pedido

```python
vendas_sk_pedidos = vendas_sk_clientes.merge(
    dim_pedidos[['id_pedido', 'sk_pedido']],
    on='id_pedido',
    how='left'
)
vendas_sk_pedidos.head(10)
```

## 5    -   Tabela dim_data

Criação da Tabela

```python
# Extração todas as datas únicas de ambos os campos
datas_unicas = pd.concat([
    dim_pedidos['data_pedido'], 
    dim_pedidos['data_envio']
], ignore_index=True).dropna().drop_duplicates().sort_values()

# Criação a dimensão de data
dim_data = pd.DataFrame({'data': datas_unicas})

# Gerando atributos derivados
dim_data['ano']           = dim_data['data'].dt.year
dim_data['mes']           = dim_data['data'].dt.month
dim_data['dia']           = dim_data['data'].dt.day
dim_data['dia_semana']    = dim_data['data'].dt.weekday + 1
dim_data['nome_dia_semana'] = dim_data['data'].dt.day_name()
dim_data['nome_mes']      = dim_data['data'].dt.month_name()
dim_data['ano_mes']       = dim_data['data'].dt.strftime('%Y-%m')
dim_data['trimestre']     = dim_data['data'].dt.quarter

# Gerando a surrogate key para cada data
dim_data['sk_data'] = range(1, len(dim_data)+1)

# Reordenando
colunas = ['sk_data', 'data', 'ano', 'mes', 'dia', 'dia_semana', 'nome_dia_semana', 'nome_mes', 'ano_mes', 'trimestre']
dim_data = dim_data[colunas]

dim_data
```

Associação a tabela dim_pedidos x sk_data_pedido

```python
dim_pedidos = dim_pedidos.merge(
    dim_data[['sk_data', 'data']],
    left_on='data_pedido',
    right_on='data',
    how='left'
).rename(columns={'sk_data': 'sk_data_pedido'}).drop(columns=['data'])

dim_pedidos
```

Padronização de datas

```python
dim_pedidos['data_pedido'] = pd.to_datetime(dim_pedidos['data_pedido']).dt.date
dim_pedidos['data_envio']  = pd.to_datetime(dim_pedidos['data_envio']).dt.date
dim_data['data'] = pd.to_datetime(dim_data['data']).dt.date
```

Associação a tabela dim_pedidos x sk_data_envio

```python
dim_pedidos = dim_pedidos.merge(
    dim_data[['sk_data', 'data']],
    left_on='data_envio',
    right_on='data',
    how='left'
).rename(columns={'sk_data': 'sk_data_envio'}).drop(columns=['data'])
dim_pedidos
```

## 7 - Outliers

IQR

```python
for coluna in ['valor_venda', 'custo_envio', 'valor_unit']:
    Q1 = fato_vendas[coluna].quantile(0.25)
    Q3 = fato_vendas[coluna].quantile(0.75)
    IQR = Q3 - Q1
    lim_inf = Q1 - 1.5 * IQR
    lim_sup = Q3 + 1.5 * IQR
    outliers = fato_vendas[(fato_vendas[coluna] < lim_inf) | (fato_vendas[coluna] > lim_sup)]
    print(f"\n--- Outliers em {coluna} ---")
    print(outliers[[coluna, 'sk_pedido', 'sk_produto']])
    print(f"Qtd de outliers em {coluna}: {outliers.shape[0]}")
```

Z-Score

```python
for coluna in ['valor_venda', 'custo_envio', 'valor_unit']:
    media = fato_vendas[coluna].mean()
    std = fato_vendas[coluna].std()
    threshold = 3  # padrão: considera outlier acima de 3 desvios
    outliers = fato_vendas[np.abs(fato_vendas[coluna] - media) > threshold * std]
    print(f"\n--- Outliers em {coluna} pelo Z-score ---")
    print(outliers[[coluna, 'sk_pedido', 'sk_produto']])
    print(f"Qtd de outliers em {coluna}: {outliers.shape[0]}")
```
